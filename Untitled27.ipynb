```python
df = df.drop(columns=['ADDRESSLINE2', 'ADDRESSLINE1', 'PHONE', 'CONTACTLASTNAME', 'CONTACTFIRSTNAME', 'POSTALCODE'])
df.shape
```

```python
df['STATE'] = df['STATE'].fillna('Unknown')
df['TERRITORY'] = df['TERRITORY'].fillna('Unknown')
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df['ORDERDATE'] = pd.to_datetime(df['ORDERDATE'])
df['ORDERDATE'].head()
```

```python
df[['QUANTITYORDERED', 'PRICEEACH', 'SALES', 'MSRP']].describe()
```

```python
monthly_sales = df.groupby(df['ORDERDATE'].dt.to_period('M'))['SALES'].sum()

plt.figure(figsize=(14,5))
monthly_sales.plot(kind='line', marker='o')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
product_sales = df.groupby('PRODUCTLINE')['SALES'].sum().sort_values(ascending=False)

plt.figure(figsize=(10,5))
product_sales.plot(kind='bar', color='steelblue')
plt.title('Total Sales by Product Line')
plt.xlabel('Product Line')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
country_sales = df.groupby('COUNTRY')['SALES'].sum().sort_values(ascending=False)

plt.figure(figsize=(12,5))
country_sales.plot(kind='bar', color='coral')
plt.title('Total Sales by Country')
plt.xlabel('Country')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
dealsize_sales = df.groupby('DEALSIZE')['SALES'].mean().sort_values(ascending=False)

plt.figure(figsize=(7,5))
dealsize_sales.plot(kind='bar', color='seagreen')
plt.title('Average Sales by Deal Size')
plt.xlabel('Deal Size')
plt.ylabel('Average Sales')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(7,6))
sns.heatmap(df[['QUANTITYORDERED', 'PRICEEACH', 'SALES', 'MSRP']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

## Key Insights

1. **Seasonality:** Sales show a clear seasonal pattern, with a sharp spike around November in both years — likely driven by holiday/year-end shopping. The business should prepare additional inventory and staffing ahead of Q4 each year.

2. **Top Product Line:** Classic Cars generate the highest total sales among all product lines, making it the company's core revenue driver. Marketing and inventory priorities should focus on this category.

3. **Top Markets:** The USA and Spain are the top two revenue-generating markets. These regions represent the company's strongest customer base and warrant continued investment.

4. **Deal Size Impact:** Average sales increase clearly from Small to Medium to Large deals (roughly 4x difference between Small and Large), confirming that deal size is a strong indicator of order value.

5. **Correlation Findings:** Unlike the customer dataset analyzed in the earlier EDA project (where correlations were near zero), this sales dataset shows strong, meaningful correlations — Sales is moderately to strongly correlated with Price (0.66), Quantity (0.55), and MSRP (0.64). This makes sense because Sales is mathematically derived from price and quantity, so a real relationship is expected here.

## Conclusion

This analysis of retail sales data reveals clear, actionable patterns: a strong holiday-season sales spike, Classic Cars as the leading product line, the USA and Spain as top markets, and a predictable relationship between deal size and revenue. Unlike the earlier customer behavior dataset, this sales data shows strong correlations between its numeric features — a reminder that the presence or absence of correlation depends heavily on the nature of the data and how its variables are related.
